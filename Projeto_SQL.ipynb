# **Projeto análise de dados com SQL.**

# **Objetivos**

Através dos dados que temos sobre os clientes de um determindado banco, estaremos fazendo uma analise para poder descobrir quais pontos podemos estar melhorando em nosso negocio, como quais as faixas salariais que mais gastam , se são homens ou mulheres e demais pontos que possamos estar verificando como interesses de business.

Nossa base de dados que será utilizada é composta por dados de creditos bancários, sendo uma fração de uma base maior.
Para verificar o conjunto completo dos dados link a seguir:
https://github.com/andre-marcos-perez/ebac-course-utils/tree/main/dataset

**Legendas das colunas.**

- idade = idade do cliente
- sexo = sexo do cliente (F ou M)
- dependentes = número de dependentes do cliente
- escolaridade = nível de escolaridade do clientes
- salario_anual = faixa salarial do cliente
- tipo_cartao = tipo de cartao do cliente
- qtd_produtos = quantidade de produtos comprados nos últimos 12 meses
- iteracoes_12m = quantidade de iterações/transacoes nos ultimos 12 meses
- meses_inativo_12m = quantidade de meses que o cliente ficou inativo
- limite_credito = limite de credito do cliente
- valor_transacoes_12m = valor das transações dos ultimos 12 meses
- qtd_transacoes_12m = quantidade de transacoes dos ultimos 12 meses

**Qual a faixa salarial e quantos clientes se encaixam em cada uma delas ?**

**Query:**

 select count(*), salario_anual from credito where salario_anual != 'na' group by salario_anual;

A maioria dos clientes se enquandram dentro da faixa salarial até 40 mil dolares e a minoria acima de 120 mil. Sendo as demais faixas salariais mais equilibradas, sem diferenças significativas de quantidades de clientes.

Ps. Para uma analise mais limpa, foram retirados os valores nulos.

**Divisão de clientes por sexo.**

**Dentro os dois qual tem o maior valor de gastos ?.**

**Query:**

select max(valor_transacoes_12m) as maior_valor_gasto, avg(valor_transacoes_12m) as media_valor_gasto, min(valor_transacoes_12m) as min_valor_gasto, sexo from credito group by sexo

WhatsApp Image 2024-05-16 at 13.47.31.jpeg

Apesar de não ser uma diferença muito grande, as mulheres tendem a ter um gasto maior que os homens.

**Com relação a faixa salarial , há correlação com o limite de credito do cliente ?.**

**Query:**

select avg(qtd_produtos) as qts_produtos, avg(valor_transacoes_12m) as media_valor_transacoes, avg(limite_credito) as media_limite, sexo, salario_anual from credito where salario_anual != 'na' group by sexo, salario_anual order by avg(valor_transacoes_12m) desc

WhatsApp Image 2024-05-16 at 13.51.58.jpeg

A resposta é sim , existe uma correlação positiva entre o salario anual e o limite de credito do cliente, onde podemos tambem notar que os homens estão na frente nesse quesito, ganhando mais que as mulheres.

**E se tem relação com os gastos mensais ?**

**Query:**

SELECT avg(valor_transacoes_12m),salario_anual from credito group by salario_anual ;

WhatsApp Image 2024-05-17 at 11.08.21.jpeg

Dois pontos devem ser destacados nessa situação , os dados que constam como "na" (nulos) podem ser pessoas que não sabem ou não querem informar suas receitas anuais, todavia são as que mais gastam mensalmente, seguidos de perto pelas pessoas com menores rendas mensais. Em contrapartida as pessoas com os maiores salarios tendem a gastar menos.

## **Conclusão da Análise de Dados**
- A maioria dos clientes, com base nesta amostra de dados, encontra-se na faixa salarial de até 40 mil dólares anuais.
- A grande maioria dos cartões utilizados são os Blue.
- Mulheres gastam levemente mais do que os homens.
Existe uma correlação entre o salário anual e o limite de crédito de cada cliente.
- Em contrapartida, quem mais gasta são os clientes com menores rendas.
## **Ações que Podem Ser Tomadas**
- Avaliar os Motivos de Menores Gastos com Outros Cartões:

- Juntando a informação de que quem mais utiliza e mais gasta são os clientes que ganham até 40 mil dólares anuais, que em sua maioria utilizam cartões Blue, é importante investigar por que os clientes com outros tipos de cartões gastam menos. Será devido a benefícios inferiores ou falta de um marketing mais agressivo? Análises detalhadas podem revelar oportunidades para melhorar a atratividade desses cartões.
- Aumentar o Ticket Médio de Gasto dos Homens:

- Estudar maneiras de aumentar o ticket médio de gasto dos homens. Embora a diferença de gasto entre homens e mulheres não seja tão grande, igualar os gastos pode gerar resultados significativos devido à maior quantidade de homens. Estratégias de marketing e promoções direcionadas podem ajudar a alcançar este objetivo.

- Foco em Clientes de Cartões Platina:

Um número muito reduzido de pessoas utiliza cartões Platina e gastam relativamente pouco se comparado aos outros cartões, mesmo pertencendo a uma faixa de clientes "premium". É crucial observar mais atentamente este grupo, pois, apesar de serem em menor quantidade, são clientes importantes em outros aspectos sociais. Incentivos e benefícios específicos podem ser desenvolvidos para aumentar o engajamento e o gasto deste segmento.
